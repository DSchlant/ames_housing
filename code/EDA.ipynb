Below I generated the median home prices for a few categorical features to get a sense for what features would be good to control for as we proceed with building a model that will help with making recommendations to homeowners for increasing the sale value.

```python
train_set['saleprice'].groupby(train_set['neighborhood']).median().sort_values(ascending=False)
```

There is a good degree of variance in median home price by neighborhood. This will be a good feature to include.

Reviewing median sale prices by type of transaction.

```python
train_set['saleprice'].groupby(train_set['sale_type']).median().sort_values(ascending=False)
```

There is a decently large range accross sale prices for homes based on the kind of sale. Another good variable to include in analysis as it may be the basis of a recommendation.
